#### Sentence & Word Tokenization in Spacy

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Dr. strange love pav bhaji of mumbai. Hulk loves chat of delhi")
```

```python
for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

Word Tokenization In nltk

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
```

```python
sent_tokenize("Dr. Strage loves pav bhaji of mumbai. Hulk loves chat of delhi")
```

```python
from nltk.tokenize import word_tokenize

word_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```
